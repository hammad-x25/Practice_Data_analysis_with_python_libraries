```python
plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st plot
plt.plot([1, 2], [3, 4])
plt.title("Left Plot")

plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd plot
plt.plot([1, 2], [4, 3])
plt.title("Right Plot")
plt.tight_layout() 
plt.show()
```

```python
#fig, axs = plt.subplots(nrows, ncols,figsize=(width,height))

fig, axs = plt.subplots(2, 2,figsize=(10,6))  # 2 rows, 2 columns  4 plots total

# Top-left plot
axs[0, 0].plot([1, 2, 3], [3, 2, 1])
axs[0, 0].set_title("Line Plot")

# Top-right plot
axs[0, 1].bar([1, 2, 3], [3, 4, 1])
axs[0, 1].set_title("Bar Plot")

# Bottom-left plot
axs[1, 0].hist([1, 2, 1, 2, 3])
axs[1, 0].set_title("Histogram")

# Bottom-right plot
axs[1, 1].scatter([1, 2, 3], [4, 5, 6])
axs[1, 1].set_title("Scatter Plot")
plt.suptitle("Comparison between different charts!")
plt.tight_layout()  # Adjust layout so titles dont overlap
plt.show()
```
