```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the dataset
tips = sns.load_dataset("tips")

# Convert columns to string before combining
tips['group'] = tips['sex'].astype(str) + ' | ' + tips['smoker'].astype(str) + ' | ' + tips['time'].astype(str)

# Count each group
count_data = tips['group'].value_counts().reset_index()
count_data.columns = ['group', 'count']

# Optional: Sort the data for cleaner x-axis
count_data = count_data.sort_values('group')

# Plot

plt.figure(figsize=(12,6))
sns.barplot(x='group', y='count', data=count_data, palette='pastel')
plt.xticks(rotation=45, ha='right')
plt.title('Count of Male/Female by Smoker Status and Time')

plt.ylabel('Count')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')

ax.set_axis_off()
plt.show()
```

```python

from mpl_toolkits.mplot3d import Axes3D


# Sample data for the 3D bar plot
x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 2, 3, 4, 5])
z = np.zeros_like(x)  # Create an array of zeros for the initial heights
dx = dy = 0.5         # Width and depth of bars

# Create a figure and a 3D axis
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# Define colors for bars
colors = ['b', 'g', 'r', 'c', 'm']

# Create the 3D bar plot
for i in range(len(x)):
    ax.bar3d(x[i], y[i], z[i], dx, dy, z[i]+i, shade=True, color=colors[i])

# Set labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.title('Beautiful 3D Bar Plot')

# Show the 3D bar plot
plt.show()
```
